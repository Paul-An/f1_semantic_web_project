```python
from pathlib import Path
import pandas as pd

from src.ie.preprocessing import normalize_text
from src.ie.ner_utils import load_ner_model, extract_entities

nlp = load_ner_model()

selected_files = [
    "seed_page_7.txt",
    "seed_page_9.txt",
    "seed_page_10.txt",
]

all_entities = []

for file_name in selected_files:
    file_path = Path("../data/processed") / file_name

    with file_path.open("r", encoding="utf-8") as file:
        text = file.read()

    normalized_text = normalize_text(text)
    entities = extract_entities(normalized_text, nlp)

    for entity in entities:
        entity["source_file"] = file_name

    all_entities.extend(entities)

entities_df = pd.DataFrame(all_entities)
entities_df.head(30)
```

```python
entities_df["label"].value_counts()
```

```python
entities_df.groupby("source_file")["label"].value_counts()
```

```python
useful_labels = ["PERSON", "ORG", "EVENT", "DATE", "GPE"]

filtered_entities_df = entities_df[entities_df["label"].isin(useful_labels)].copy()
filtered_entities_df.head(30)
```

```python
filtered_entities_df.groupby("source_file")["label"].value_counts()
```

```python
from src.ie.entity_mapping import map_entity_to_f1_type

filtered_entities_df["f1_type"] = filtered_entities_df.apply(
    lambda row: map_entity_to_f1_type(row["text"], row["label"]),
    axis=1,
)

filtered_entities_df[["text", "label", "f1_type", "source_file"]].head(30)
```

```python
from pathlib import Path
import pandas as pd

from src.ie.preprocessing import normalize_text
from src.ie.ner_utils import load_ner_model
from src.ie.relation_candidates import extract_relation_candidates

nlp = load_ner_model()

sample_path = Path("../data/processed/seed_page_9.txt")
with sample_path.open("r", encoding="utf-8") as file:
    text = file.read()

normalized_text = normalize_text(text)

relation_candidates = extract_relation_candidates(normalized_text, nlp)
relation_candidates_df = pd.DataFrame(relation_candidates)

relation_candidates_df.head(20)
```

```python
relation_candidates_df[["entity_1_type", "entity_2_type"]].value_counts()
```

```python
from src.ie.relation_mapping import infer_f1_relation

relation_candidates_df["relation_type"] = relation_candidates_df.apply(
    lambda row: infer_f1_relation(
        row["sentence"],
        row["entity_1"],
        row["entity_1_type"],
        row["entity_2"],
        row["entity_2_type"],
    ),
    axis=1,
)

relation_candidates_df[
    ["entity_1", "entity_1_type", "entity_2", "entity_2_type", "relation_type"]
]
```

```python
relation_candidates_df["relation_type"].value_counts(dropna=False)
```

```python
mapped_relations_df = relation_candidates_df.dropna(subset=["relation_type"]).copy()
mapped_relations_df[
    ["sentence", "entity_1", "entity_1_type", "entity_2", "entity_2_type", "relation_type"]
]
```

```python
unique_relations_df = (
    mapped_relations_df[
        ["entity_1", "entity_1_type", "entity_2", "entity_2_type", "relation_type"]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)

unique_relations_df
```

```python
unique_relations_df["relation_type"].value_counts()
```

```python
from src.ie.relation_normalization import normalize_relation_direction

normalized_relations = unique_relations_df.apply(
    lambda row: normalize_relation_direction(
        row["entity_1"],
        row["entity_1_type"],
        row["entity_2"],
        row["entity_2_type"],
        row["relation_type"],
    ),
    axis=1,
)

normalized_relations = [r for r in normalized_relations if r is not None]
normalized_relations_df = pd.DataFrame(normalized_relations).drop_duplicates().reset_index(drop=True)

normalized_relations_df
```

```python
normalized_relations_df["predicate"].value_counts()
```
